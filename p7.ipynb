```python
def f(x, y):
    return (10 * x) ** 2 + (y + 2) ** 2

def f_gd(x, y, lr, itr):
    """
    Gradient descent for f(x, y)
    x, y: start point
    lr: learning rate
    itr: number of iterations
    """
    print('f value: {}\n'.format(f(x, y)))
    for i in range(itr):
        grad_x = 200 * x
        grad_y = 2 * y + 4
        
        x = x - lr * grad_x
        y = y - lr * grad_y
        
        print('Iteration: {}'.format(i + 1))
        print('x: {}, y: {}'.format(x, y))
        print('f value: {}\n'.format(f(x, y)))
```

```python
def g(x, y):
    return (x - 10) ** 2 + (x - y + 5) ** 4

def g_gd(x, y, lr, itr):
    """
    Gradient descent for g(x, y)
    x, y: start point
    lr: learning rate
    itr: number of iterations
    """
    print('g value: {}\n'.format(g(x, y)))
    for i in range(itr):
        grad_x = 2 * x - 20 + 4 * (x - y + 5) ** 3
        grad_y = 4 * -1 * (x - y + 5)
        
        x = x - lr * grad_x
        y = y - lr * grad_y
        
        print('Iteration: {}'.format(i + 1))
        print('x: {}, y: {}'.format(x, y))
        print('g value: {}\n'.format(g(x, y)))
```
